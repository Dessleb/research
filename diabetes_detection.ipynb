```python
df = pd.get_dummies(df,columns=['gender','smoking_history'],drop_first=True)
```

```python
x=df.drop('diabetes',axis=1)
y = df['diabetes']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
df.describe()
```

```python
df['diabetes'].mean()
```

feature scaling


```python
from sklearn.preprocessing import StandardScaler
```

```python
cols_to_scale = ['age','bmi','HbA1c_level','blood_glucose_level']
scaler = StandardScaler()
```

```python
X_train.loc[:,cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test.loc[:,cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

class balancing (SMOTE)


```python
from imblearn.over_sampling import SMOTE
```

```python
print(f'Diabetes cases BEFORE SMOTE: {y_train.value_counts()}')
```

```python
smote = SMOTE(sampling_strategy='minority',random_state=42)
```

```python
X_train_resampled,y_train_resampled = smote.fit_resample(X_train,y_train)
```

```python
from collections import Counter
```

model training


```python
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.metrics import classification_report
```

Random Forest


```python
'---Training Random Forest ---'
rf_model = rfc(n_estimators=200,random_state=42,n_jobs=-1)
rf_model.fit(X_train_resampled,y_train_resampled)
rf_pred = rf_model.predict(X_test)
print('∖nRandom Forest Report:')
df_report_rf = (pd.DataFrame(classification_report(y_test,rf_pred,output_dict=True)).transpose()).round(4)
```

XGBoost


```python
'---Training XGBoost---'
xgb_model = xgb.XGBClassifier(objective='binary:logistic',n_estimators=100,learning_rate=0.1,random_state=42,use_label_encoder=False,eval_metric='logloss',n_jobs=-1)
xgb_model.fit(X_train_resampled,y_train_resampled)
xgb_pred = xgb_model.predict(X_test)
'---XGB Report---'
df_report_xgb = (pd.DataFrame(classification_report(y_test,xgb_pred,output_dict=True)).transpose()).round(4)
```

svm model


```python
svm_model = SVC(kernel='rbf',random_state=42, probability=True)
svm_model.fit(X_train_resampled,y_train_resampled)
svm_pred=svm_model.predict(X_test)
'---SVM Report---'
df_report_svm = (pd.DataFrame(classification_report(y_test,svm_pred,output_dict=True)).transpose()).round(4)
```
